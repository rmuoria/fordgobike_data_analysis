## Preliminary Wrangling


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# loading and sampling the dataframe
df_gb = pd.read_csv('fordgobike.csv')
df_gb
```

```python
# high-level overview of data shape and composition
print(df_gb.shape)
print(df_gb.dtypes)
print(df_gb.head(10))
```

```python
# Converting bike_share_for_all_trip, member_gender, user_type into categorical datatype
ordinal_var_dict = {'user_type': ['Customer','Subscriber'],
                    'member_gender': ['Male', 'Other', 'Female'],
                    'bike_share_for_all_trip': ['Yes', 'No']}

for var in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ordinal_var_dict[var])
    df_gb[var] = df_gb[var].astype(ordered_var)
```

```python
#change the data type of  start_time  & end_time to datetime64
df_gb = df_gb.astype({"start_time":'datetime64', "end_time":'datetime64'})
```

```python
# checking on changes in the structure
df_gb.info()
```

```python
# descriptive statistics for numeric variables
df_gb.describe()
```

```python
# splitting start and end times into days and hours for use in our analysis
df_gb['start_time'] = pd.to_datetime(df_gb['start_time'])
df_gb['end_time'] = pd.to_datetime(df_gb['end_time'])
df_gb['start_day'] = df_gb['start_time'].dt.day_name()
#df_gb['start_month'] = df_gb['start_time'].dt.month_name()
#df_gb['start_year'] = df_gb['start_time'].dt.year
df_gb['start_hr'] = df_gb['start_time'].dt.hour
df_gb['start_hr'] = df_gb.start_hr.astype(str)

df_gb['end_day'] = df_gb['end_time'].dt.day_name()
df_gb['end_hr'] = df_gb['end_time'].dt.hour
df_gb['end_hr'] = df_gb.start_hr.astype(str)
```

## Univariate Exploration

```python
# distribution of duration of the rides
binsize = 50
bins = np.arange(50, df_gb['duration_sec'].max()+binsize, binsize)
plt.figure(figsize=[8, 5])
plt.hist(data = df_gb, x = 'duration_sec', bins=bins)
plt.title('Trip Distribution in Seconds')
plt.xlabel('Duration (Sec)')
plt.xlim([0, 10000]);
```

```python
# There is a long tail in the distribution, we need to put it on a log scale
log_binsize = 0.025
bins = 10 ** np.arange(1.5, np.log10(df_gb['duration_sec'].max())+log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df_gb, x = 'duration_sec', bins=bins)
plt.xscale('log')
plt.xticks([50, 1e2, 5e2, 1e3, 5e3, 1e4, 2e4, 5e4, 1e5], [50, '100', '500', '1k', '5k', '10k', '20k', '50k', '100k'])
plt.title('Trip Distribution in Seconds')
plt.xlabel('Duration in Seconds')
plt.ylabel('Number of Rides');
```

Duration has a longtail distribution hence the use of log to focus in more on the distribution. After using the log-scale, the distibution seems normal with higher frequecies being in the middle of the curve indicating a continous distribution. 


```python
# Distribution of trips within the days of the week
#first we arrange the days of the week and hours in order
ordinal_var_dict = {'start_day': ['Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday'],
                    'end_day': ['Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday'],
                   'start_hr': [str(i) for i in range(0,24)],
                   'end_hr': [str(i) for i in range(0,24)]}

for var in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ordinal_var_dict[var])
    df_gb[var] = df_gb[var].astype(ordered_var)
```

```python
# Distribution of bike rides per days of the week
fig, ax = plt.subplots(2, figsize = [12,10])
default_color = sb.color_palette()[0]
sb.countplot(data = df_gb, x = 'start_day', color = default_color,  ax = ax[0])
plt.title('Trip distribution by Day of Week')
sb.countplot(data = df_gb, x = 'end_day', color = default_color,  ax = ax[1]);
```

Most bike trips happened on Thursdays followed by Tuesdays. Wednesday, Friday and Monday seem to be close in terms of bike trips. Saturday and Sunday has the lowest number of bike trips. 

```python
# Distribution of bike rides per hour of the day
fig, ax = plt.subplots(2,figsize = [12,10])
default_color = sb.color_palette()[0]
sb.countplot(data = df_gb, x = 'start_hr', color = default_color,  ax = ax[0])
plt.title('Distribution of Rides per Hour of Day')
sb.countplot(data = df_gb, x = 'end_hr', color = default_color, ax = ax[1]);
```

The highest recorded trip times are in the 8th - 9th and 17th - 18th hour of the day. The lowest activity happens between the hours of 0000hrs - 0500hrs. 

```python
plt.figure(figsize = [15, 5])
sb.countplot(x="start_station_id", data=df_gb,
              order=df_gb.start_station_id.value_counts().iloc[:10].index)
plt.title('Top 10 Distribution of Start Stations')
plt.xlabel('Start Station ID')
plt.ylabel('Number of Rides');
```

The most popular start station is station 58 and 67 indicating that most people could be staying around this station. The observation shows the top 10 most popular stations.

```python
plt.figure(figsize = [15, 5])
sb.countplot(x="end_station_id", data=df_gb,
              order=df_gb.end_station_id.value_counts().iloc[:10].index)
plt.title('Top 10 Distribution of End Stations')
plt.xlabel('End Station ID')
plt.ylabel('Number of Rides');
```

Station 67 and 58 has the most number of users as the end station. This would indicate that more people stay around these two stations as they also are the most popular start stations.

```python
# Distribution of age for bike users
binsize = 1
bins = np.arange(0, df_gb['member_birth_year'].astype(float).max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df_gb.dropna(), x = 'member_birth_year', bins = bins)
# Avoiding outliners by selecting from years 1939 and grouping by 10 years
plt.axis([1939, 2009, 0, 12000])
plt.xticks([1939, 1949, 1959, 1969, 1979, 1989, 1999, 2009], [(2019-1939), (2019-1949), (2019-1959), (2019-1969), (2019-1979), (2019-1989), (2019-1999), (2019-2009)])
plt.gca().invert_xaxis()
plt.title('Distribution of User Age')
plt.xlabel('Age in Years')
plt.ylabel('Number of Rides')
plt.show()
```

The distribution users is concentrated between 20-30 after which we notice a steady decline from around 35 years moving forward.

```python
# plotting types of users on a pie chart.
user_type = df_gb['user_type'].value_counts()
plt.pie(user_type, explode =  (0, 0.1), labels = user_type.index, shadow = True, startangle = 90,
        counterclock = False, autopct='%1.1f%%');
plt.title('Distribution of User Types')
plt.axis('square');
df_gb['user_type'].value_counts()

```

The distribution shows that most users are subscribers at almost 90% while the rest are customers at 10%. 

```python
# Next we look at the distribution of genders using the bike ride system
sb.countplot(data = df_gb, x = 'member_gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Rides');
df_gb['member_gender'].value_counts()
```

The distibution shows a large number of users are male with over 130,000 user vs females at a little over 40,000.

```python
# Finally we look at the distibution of the bike share for all program
ax = sb.countplot(data = df_gb, x = 'bike_share_for_all_trip',  color=sb.color_palette()[0],
                  order=df_gb.bike_share_for_all_trip.value_counts().index)
total = len(df_gb)
for p in ax.patches:
    percentage = '{:.3f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()/2
    y = p.get_height()+.05
    ax.annotate(percentage, (x, y),ha='center')
plt.title('Bike Share Program Distribution')
plt.xlabel('Bike Share for All')
plt.ylabel('Number of Rides');
df_gb.bike_share_for_all_trip.value_counts()
```

The distribution indicates that most users (90%) do not use the share for all program. Only about 10% make use of the program.

## Bivariate Exploration


```python
# We first look at the correlation between duration and user age
from datetime import datetime
df_gb['age'] = datetime.now().year - df_gb.member_birth_year

plt.figure(figsize=[8,5])
plt.scatter(data = df_gb, x = 'age', y ='duration_sec', alpha= None)
plt.title('Trip distribution by Age')
plt.xlabel('Age')
plt.ylabel('Trip Duration in Seconds')
plt.show()
```

The correlation is high between the ages of between 25 and 35. The data points are highly concentrated around this area.

```python
# Next we look at the correlation betweeen trip duration and gender
plt.figure(figsize = [8, 5])
base_color = sb.color_palette()[1]
sb.boxplot(data = df_gb, x = 'member_gender', y = 'duration_sec', color = base_color)
plt.ylim([-10, 4000])
plt.title('Trip Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Duration in Seconds')
plt.show()
```

This observation shows that the females take a slighlty longer time the males for the rides. 

```python
# We not look at the correlation between user type and duration
plt.figure(figsize = [16, 5])
base_color = sb.color_palette()[0]
sb.violinplot(data=df_gb, x='user_type', y='duration_sec', color=base_color, inner='quartile')
plt.ylim([-10, 6000])
plt.xticks(rotation=15)
plt.title('Trip distribution by User Type')
plt.xlabel('User Type')
plt.ylabel('Duration in Seconds')
plt.show();
```

This obervation indicates that subscribers are spending more time on their rides, and most are using about 1000 seconds between their start and end points compared the customers.

```python
# Lastly we look as the correlation between user type and the  bike for all share program

sb.countplot(data = df_gb, x = 'user_type', hue = 'bike_share_for_all_trip', order=df_gb.user_type.value_counts().index,
             palette = 'Blues')
plt.legend(loc = 1, ncol = 2)
plt.legend(title='Bike Share Prog')
plt.title('Trip Distribution by User Type and Bike Share Program')
plt.xlabel('User Type')
plt.ylabel('Duration in Seconds')
plt.show();
```

From the observation, we can conclude that there is a correlation between subscibers and the bike share program, while no correlation exists between customers and the program.

## Multivariate Exploration

```python
# Let us observe the duration across day of week and user type
fig = plt.figure(figsize = [18,6])
ax = sb.pointplot(data = df_gb, x ='start_day', y = 'duration_sec', hue = 'user_type',
           palette = 'Blues', linestyles = '', dodge = 0.4)
plt.title('Trip Duration Across Days of the Week and User Type')
plt.xlabel('User Type')
plt.ylabel('Trip Duration in Seconds')
plt.legend(title='User Type')
ax.set_yticklabels([],minor = True)
plt.show();
```

The observation shows that customers take more time on their rides on the weekdays compared to the subscribers. The trips for customers are even longer on the weekends. Subscribers also take longer trips on the weekends but still not as long as the customers.

```python
# Now let us observe the duration across gender and bike share program
fig = plt.figure(figsize = [10,6])
ax = sb.pointplot(data = df_gb, x ='member_gender', y = 'duration_sec', hue = 'bike_share_for_all_trip',
           palette = 'Blues', linestyles = '', dodge = 0.4)
plt.title('Trip Duration Across Gender and Bike Sharing Program')
plt.xlabel('Gender')
plt.ylabel('Trip Duration in Seconds')
plt.legend(title='Bike Share Program')
ax.set_yticklabels([],minor = True)
plt.show();
```

There are shorter trips done by males whether in the bike sharing program or not compared to the other genders. Also both males and females who are not in the bike share program seem to take shorter trips. On the contrary, the other gender who are not in the bike sharing program seem takes more time on their trips.

Further analysis strengthens the previous results where we can observe that subscribers are taking shorter ride times on the weekdays ccompare to the weekends and also males are taking less time between their destinations.
